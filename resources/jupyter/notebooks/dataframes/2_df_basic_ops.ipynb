```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('ops').getOrCreate()
```

```python
# CSV has option to inferSchema which json doesnt have
df = spark.read.csv('../data/appl_stock.csv', inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.head(3)[0]
```

```python
# Filter operation using dataframes
df.filter('Close < 500').show()
```

```python
# Combine multiple fuctions
df.filter('Close < 500').select(['open', 'close']).show()
```

```python
# Preferable syntax
df.filter(df['close'] < 500).select(['open', 'close']).show()
```

```python
# Filtering on multiple conditions
# Common Errors:
# 1. Use of "and", "or" rather than "&", "|"
# ex : (error will be thrown)
df.filter(df['close'] < 200 and df['open'] > 100).show()
# 2. Encapsulation of conditions within the parenthesis
# ex : (even after correction the operators it will throw an error due to parenthesis)
df.filter(df['close'] < 200 & df['open'] > 100).show()
# Correct syntax:
df.filter( (df['close'] < 200) & (df['open'] > 100) ).show()
```
